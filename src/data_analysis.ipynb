```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler
%matplotlib inline

np.random.seed(0)

leaf_data = pd.read_csv("../data/train.csv")
leaf_data.head()
```

```python
ids = leaf_data['id'].values
target = leaf_data['species'].values
le = LabelEncoder()
target = le.fit_transform(target)
classes = le.classes_

unique, counts = np.unique(target,return_counts=True)
print('max: ',counts.max())
print('min: ',counts.min())
```

```python
data = leaf_data.drop(['id','species'],1).values
# maybe I need to scale each sub feature set separately?
data = StandardScaler().fit_transform(data) 
```

```python
from sklearn.ensemble import RandomForestClassifier

# default 10 estimators
clf = RandomForestClassifier(n_estimators = 100)
clf.fit(data,target)
```

```python
leaf_test = pd.read_csv("../data/test.csv")
leaf_test.head()
```

```python
test_id = leaf_test['id'].values
test = leaf_test.drop('id',1).values
test = StandardScaler().fit_transform(test)

# totally blind no validation or anything
predictions = clf.predict_proba(test)
```

```python
submission = pd.DataFrame(predictions, columns=classes)
submission.insert(0,'id',test_id)
submission.head()
```

```python
submission.to_csv('../submission/submission.csv',index=False)
```
